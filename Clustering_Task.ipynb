```python
def remove_outliers(num_col):
    for col in num_col.select_dtypes(include=[np.number]).columns:
        q1 = num_col[col].quantile(0.25)
        q3 = num_col[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        num_col = num_col[(num_col[col] >= lower_bound) & (num_col[col] <= upper_bound)]
    return num_col

data_cleaned = remove_outliers(num_col)
```

```python
imputer = SimpleImputer(strategy='mean')
data_imputed = pd.DataFrame(imputer.fit_transform(data_cleaned), columns=data_cleaned.columns)
```
